```python
# file: app.py
import ccxt
import pandas as pd
import plotly.graph_objects as go
import streamlit as st

exchange = ccxt.binance()

def fetch_data(tf='1h'):
    ohlcv = exchange.fetch_ohlcv('BTC/USDT', timeframe=tf, limit=300)
    df = pd.DataFrame(ohlcv, columns=["timestamp", "open", "high", "low", "close", "volume"])
    df["timestamp"] = pd.to_datetime(df["timestamp"], unit="ms")
    return df

st.title("BTC/USDT Chart")

timeframe = st.selectbox("Select Timeframe", [
    "1h", "2h", "3h", "4h", "6h", "12h",
    "1d", "2d", "3d", "4d", "5d", "6d",
    "1w", "2w", "3w", "4w",
    "1M", "2M", "3M", "4M"
])

df = fetch_data(tf=timeframe)

fig = go.Figure(data=[
    go.Candlestick(
        x=df["timestamp"],
        open=df["open"],
        high=df["high"],
        low=df["low"],
        close=df["close"]
    )
])

fig.update_layout(title=f"BTC/USDT ({timeframe})", xaxis_title="Time", yaxis_title="Price")
st.plotly_chart(fig, use_container_width=True)
```
